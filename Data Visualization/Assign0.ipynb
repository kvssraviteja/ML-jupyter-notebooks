#### Sample 1 : Regression

For regression, a data set describes the temperatures and humidities in various rooms of a house, located in belgium and the energy consumtion of various appliances in the house. The data also presents the outside temperature and humidity which is derived from the nearest airport weather station. Two random variables have also been included in the data set for testing the regression models and to filter out the non predictive attributes. The data is collected for every 10 min over a period of 4.5 months. 

Attributes:
date, applicances: denotes the energy usage of appliacnes in Whr, lights: energy usage of lights in Whr, Temperatures and humidty in various rooms, temperature and humidity outside, pressure, visibility, Temperature dew point and 2 random variables.

#### Sample 2: Classification

For classification, the NDVI(Normalised difference vegetation index) values of various classes of lands are used as data. NDVI index denotes the graphical representation to asses an area's green vegetation. The data set was derived using sources: 1)  Landsat time-series satellite imagery from the years 2014-2015, 2) crowdsourced georeferenced polygons with land cover labels obtained from OpenStreetMap. 5 classes of lands are recorded: farm, forest, water, orchard and grass. The data set also has some samples whose land class is not known. 

Attributes:
Class, maximum NDVI for the sample, 28 recordings of NVDI from period of January 2014 to July 2015.  

```python
data1 = pd.read_csv("training.csv")
data1
```

There are some rows with unknown class type and they are denoted as "impervious". We can clean the data by removing them, so that we can work on data whose class type is known.

```python
data1 = data1[data1["class"] != 'impervious']
data1
```

```python
count = pd.value_counts(data1['class'].values, sort=False)
plt.plot(count, "*")
```

The above plot shows the number of samples taken for each class of land. Let's look at the maximum NDVI values in each class of land. 

```python
plt.plot(data1["class"],data1["max_ndvi"],".")
```

It can be observed that the maximum NVDI for forest is high than any other class of land. We can also observe that most of the samples of orchard, have very high NVDI. Water have very low NVDI samples compared to other classes. 

```python
fig = plt.figure(figsize=(20,60))
plt.clf()
for i in range(2,29):
    plt.subplot(15, 2, i+1)
    plt.plot(data1["class"],data1.iloc[:,i],".")
    plt.ylabel(data1.columns.values[i])
fig.tight_layout()

```

Considering the samples from various time periods, maximum variation can be observed in Orchards. Sometimes the NVDI is very low and sometimes it very high. Farms and Forests have very constant NVDI throughout the observed time period.
